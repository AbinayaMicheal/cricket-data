## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'overs' and 'ballnumber' to a single sortable metric, 'delivery_count'
# Assuming each over has 6 balls (simplified version, ignoring extras like wides or no-balls)
df['delivery_count'] = df['overs'] * 6 + df['ballnumber']

# Sort DataFrame by this new 'delivery_count' column
df.sort_values(by='delivery_count', inplace=True)

# Now plot 'total_run' as a function of 'delivery_count' to visualize run progression over deliveries
plt.plot(df['delivery_count'], df['total_run'], color='blue', marker='o', linestyle='-')
plt.xlabel('Delivery Count')
plt.ylabel('Total Runs')
plt.title('Progression of Total Runs Over Deliveries')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
